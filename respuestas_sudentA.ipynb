### **Problema 1: Métodos de Intervalo vs. Métodos Abiertos**  

**1. Garantía de convergencia**  
- **Métodos de intervalo (Bisección)**:  
  - Convergencia garantizada si \( f \) es continua en \([a, b]\) y \( f(a) \cdot f(b) < 0 \).  
- **Métodos abiertos (Newton-Raphson)**:  
  - No hay garantía; depende de \( x_0 \) y la naturaleza de \( f \).  

**2. Velocidad de convergencia**  
- **Bisección**: Lineal (\( \epsilon_{k+1} \approx \frac{1}{2} \epsilon_k \)).  
- **Newton-Raphson**: Cuadrática (\( \epsilon_{k+1} \approx C \epsilon_k^2 \)) bajo condiciones ideales.  

**3. Requerimientos de información**  
- **Bisección**: Solo \( f(x) \) (no necesita derivadas).  
- **Newton-Raphson**: Requiere \( f(x) \) y \( f'(x) \).  

**4. Estabilidad numérica**  
- **Bisección**: Muy estable (solo requiere continuidad).  
- **Newton-Raphson**: Inestable si \( f'(x) \approx 0 \) o \( x_0 \) está lejos de la raíz.

### **Problema 2: Newton-Raphson para \( f(x) = e^{-x} - \ln(x) \)**  

**1. Fórmula de iteración**  
\[
x_{k+1} = x_k - \frac{e^{-x_k} - \ln(x_k)}{-e^{-x_k} - \frac{1}{x_k}}
\]  

**2. Dos iteraciones con \( x_0 = 1.5 \)**  

| Iteración | \( x_k \)   | \( f(x_k) \)       | \( f'(x_k) \)      | \( x_{k+1} \) | Error relativo (%) |
|-----------|------------|--------------------|--------------------|---------------|--------------------|
| 0         | 1.5        | \(-0.1824\)        | \(-0.8898\)        | 1.2950        | \(-\)              |
| 1         | 1.2950     | \(0.0150\)         | \(-1.0461\)        | 1.3093        | 15.83%             |
| 2         | 1.3093     | \(-0.0002\)        | \(-1.0416\)        | 1.3098        | 1.09%              |

**3. ¿Por qué no converge si \( x_0 \approx 0 \)?**  
- \( \ln(x) \to -\infty \) y \( f'(x) \to -\infty \), causando división por cero o overflow.

### **Problema 3: Convergencia de Newton-Raphson**  

**1. ¿Siempre converge cuadráticamente?**  
❌ **No**. La convergencia cuadrática requiere:  
   - \( f \) sea \( C^2 \) cerca de la raíz.  
   - \( f'(\alpha) \neq 0 \) (raíz simple).  

**2. Condiciones para convergencia cuadrática**  
- \( f \in C^2 \) en un entorno de la raíz \( \alpha \).  
- \( x_0 \) suficientemente cercano a \( \alpha \).  

**3. Ejemplo de no convergencia**  
- Función: \( f(x) = x^{1/3} \), \( x_0 = 1 \).  
- Las iteraciones oscilan: \( x_{k+1} = -2x_k \).

### **Problema 4: Comparación de Bisección y Newton-Raphson**  

**1. Bisección: Número mínimo de iteraciones**  
\[
\frac{3 - 2}{2^n} < 0.001 \implies n > \log_2(1000) \approx 9.96 \implies n_{\text{mín}} = 10.
\]  

**2. Newton-Raphson: 1 iteración con \( x_0 = 2.5 \)**  
\[
f(2.5) = 5.625, \quad f'(2.5) = 16.75 \implies x_1 = 2.5 - \frac{5.625}{16.75} \approx 2.1642.
\]  

**3. Eficiencia comparada**  

| Método          | Evaluaciones por iteración | Convergencia | Iteraciones típicas |
|-----------------|---------------------------|--------------|---------------------|
| Bisección       | 1 (\( f \))               | Lineal       | Alta (\(\sim 10\))  |
| Newton-Raphson  | 2 (\( f \) y \( f' \))    | Cuadrática   | Baja (\(\sim 3-5\))  |

---  

**Conclusión**:  
- **Bisección** es lento pero seguro.  
- **Newton-Raphson** es rápido pero sensible a \( x_0 \).
